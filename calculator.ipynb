```python
from tkinter import *
import math

root = Tk()
root.geometry("450x600")
root.title("Calculator")
root.resizable(width=False, height=False)
root.configure(bg="lightgray")

# Entry widget for displaying input and results
text_input = StringVar()
entry_display = Entry(root, font=('arial', 20), textvariable=text_input, bd=15, insertwidth=4, width=20,
                      justify='right', bg="powder blue")
entry_display.grid(row=0, column=0, columnspan=4, pady=10)

# Function to update the input display
def show(value):
    current_input = text_input.get()
    text_input.set(current_input + str(value))

# Function to perform calculations
def calculate():
    try:
        result = eval(text_input.get())
        text_input.set(result)
    except Exception as e:
        text_input.set("Error")

# Function to clear the input display
def clear():
    text_input.set("")

# Buttons
buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    '.', '0', '=', '+'
]

row_val = 1
col_val = 0

for button in buttons:
    Button(root, text=button, padx=20, pady=20, font=('arial', 16, 'bold'),
           command=lambda b=button: show(b) if b != '=' else calculate(),
           bg="lightblue" if button.isdigit() or button == '.' else "lightgreen").grid(row=row_val, column=col_val, padx=5, pady=5)
    col_val += 1
    if col_val > 3:
        col_val = 0
        row_val += 1

# Additional functions buttons
Button(root, text='C', padx=20, pady=20, font=('arial', 16, 'bold'), command=clear, bg="orange").grid(row=row_val, column=col_val, padx=5, pady=5)
col_val += 1
Button(root, text='âˆš', padx=20, pady=20, font=('arial', 16, 'bold'), command=lambda: show('math.sqrt('), bg="lightgreen").grid(row=row_val, column=col_val, padx=5, pady=5)
col_val += 1
Button(root, text='x^2', padx=20, pady=20, font=('arial', 16, 'bold'), command=lambda: show('**2'), bg="lightgreen").grid(row=row_val, column=col_val, padx=5, pady=5)
col_val += 1
Button(root, text='(', padx=20, pady=20, font=('arial', 16, 'bold'), command=lambda: show('('), bg="lightgreen").grid(row=row_val, column=col_val, padx=5, pady=5)
col_val += 1
Button(root, text=')', padx=20, pady=20, font=('arial', 16, 'bold'), command=lambda: show(')'), bg="lightgreen").grid(row=row_val, column=col_val, padx=5, pady=5)

root.mainloop()
```
